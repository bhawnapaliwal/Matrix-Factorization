# Non-Negative Matrix Factorization (NMF)

```python
#non negative matrix factorization part1:
u_mat4 = np.random.rand(rows,k)
v_mat4 = np.random.rand(k,cols)
u_mat4[np.isnan(u_mat4)] = np.random.randn(len(u_mat4[np.isnan(u_mat4)]))
v_mat4[np.isnan(v_mat4)] = np.random.randn(len(v_mat4[np.isnan(v_mat4)]))
w_mat = np.dot(np.transpose(r_m),r_m)
l = []
for i in range(cols):
    s=0
    for j in range(cols):
        s += w_mat[i][j]
    l.append(s)
d_mat = np.diag(l)
```

```python
#non negative matrix factorization part2:
import math
import copy
for _iter in range(100):
#     print(r_m.shape)
#     print(u_mat4.shape)
#     print(v_mat4.shape)
    error_mat = r_m-np.dot(u_mat4,v_mat4)
    for i in range(rows):
        for j in range(cols):
            if math.isnan(error_mat[i][j]):
                print(str(i)+' '+str(j))
#     print(error_mat.shape)
#     if _iter%10==0:
    print('iter: '+str(_iter))
    print(np.linalg.norm(error_mat,ord='fro'))
    den1 = copy.deepcopy(np.dot(np.dot(u_mat4,v_mat4),np.transpose(v_mat4)))
    for i in range(rows):
        for j in range(k):
            if den1[i][j]==0:
                den1[i][j]=1e-10
    u_mat_temp = copy.deepcopy(u_mat4*((np.dot(r_m,np.transpose(v_mat4)))/den1))
    u_mat4 = copy.deepcopy(u_mat_temp)
    den2 = copy.deepcopy(np.dot(np.dot(np.transpose(u_mat4),u_mat4),v_mat4)+_lambda*np.dot(v_mat4,d_mat))
    for i in range(k):
        for j in range(cols):
            if den2[i][j]==0:
                den2[i][j]=1e-10
    v_mat_temp = copy.deepcopy(v_mat4*((np.dot(np.transpose(u_mat4),r_m) + _lambda*np.dot(v_mat4,w_mat))/den2))
    v_mat4 = copy.deepcopy(v_mat_temp)
```

```python
temp_mat = np.dot(u_mat4,v_mat4)
s=0
cnt=0
for i in range(rows):
    for j in range(cols):
        if test_mat[i][j]>0:
            x = test_mat[i][j]-temp_mat[i][j]
            x = x*x
            s += x
            cnt+=1
rmse_error = math.sqrt(s/cnt)
```
